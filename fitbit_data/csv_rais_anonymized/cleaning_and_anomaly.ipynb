Anomaly detection method 1 : Using pdfs, we can use a threshold value to identify the observations with low probabilities

```python
import numpy as np
from scipy.stats import norm


# Compute the mean and standard deviation of each feature
means = np.mean(df3, axis=0)
stds = np.std(df3, axis=0)

# Compute the probability density function of each observation
pdfs = np.prod(norm.pdf((df3 - means) / stds), axis=1)

# Identify the observations with low probabilities as anomalies
anomalies = df3[pdfs < 0.0000000001]

# Print the number of anomalies and the indices of the anomalous observations
# print(f'Number of anomalies {anomalies}')
```

Anomaly Detection method 2 : K-means which partitions the data into K clusters based on their similarity.

```python
import numpy as np
from sklearn.cluster import KMeans


# Fit a K-means clustering model to the data
kmeans = KMeans(n_clusters=2)
kmeans.fit(df3)

# Assign each observation to a cluster
labels = kmeans.labels_

# Print the number of observations in each cluster
print(f'Number of observations in cluster 0: {np.sum(labels == 0)}')
print(f'Number of observations in cluster 1: {np.sum(labels == 1)}')
```

Anomaly Detection method 3 : GMM 
 Gaussian mixture model (GMM) to model the underlying distribution of the data, and then identify the observations that are far from the model.

```python
import numpy as np
from sklearn.mixture import GaussianMixture


# Fit a Gaussian mixture model to the data
gmm = GaussianMixture(n_components=3)
gmm.fit(df3)

# Compute the log-likelihood scores for each observation
log_probs = gmm.score_samples(df3)

# Identify the observations with low log-likelihood scores as anomalies
anomalies = df3[log_probs < -35]  #was originally -3

# Print the number of anomalies and the indices of the anomalous observations
print(f'Number of anomalies: {anomalies.shape[0]}')
print(f'Anomalous indices: {np.where(log_probs < -3)[0]}')
```

Anomaly Detection method 4: Isolation Forest

```python
from sklearn.ensemble import IsolationForest


# Select columns for anomaly detection
X = df3[['calories', 'sedentary_minutes', 'very_active_minutes']]

# Create the isolation forest model
isolation_forest = IsolationForest(n_estimators=100, contamination='auto', random_state=42)

# Fit the model to the data
isolation_forest.fit(X)

# Predict the anomalies
isolation_predictions = isolation_forest.predict(X)

# Add the predictions to the original dataset
df['isolation_predictions'] = isolation_predictions

# Count the number of anomalies detected
num_anomalies = len(df[df['isolation_predictions'] == -1])

# Print the number of anomalies detected
print('Number of anomalies detected:', num_anomalies)
```

Anomaly Detection on a Target variable : fitting a Gaussian distribution to the data and identifying points that fall outside a certain probability threshold.

```python
import numpy as np
from scipy.stats import norm



# Select column for anomaly detection
X = df3['calories']

# Fit a Gaussian distribution to the data
mu, std = norm.fit(X)

# Set the probability threshold
prob_threshold = 0.00001

# Calculate the probability of each data point
probs = norm.pdf(X, mu, std)

# Identify the anomalies using the probability threshold
anomalies = X[probs < prob_threshold]

# Print the anomalies
print('Anomalies:', anomalies)
```
